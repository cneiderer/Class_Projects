```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud
import spacy
```

```python
df = pd.read_csv('/home/cneiderer/Downloads/inaug_speeches_mod.csv')
```

```python
df.head()
```

```python
from gensim import corpora, models, similarities, matutils
# sklearn
from sklearn import datasets
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans
```

```python
docs = list(df.text)
len(docs)
```

```python
nlp = spacy.load('en')
```

```python
doc = nlp(docs[0])
```

```python
# Create a CountVectorizer for parsing/counting words
count_vectorizer = CountVectorizer(ngram_range=(3, 4), min_df=0.02, max_df=0.8,
                                   stop_words='english', token_pattern="\\b[a-z][a-z]+\\b")
count_vectorizer.fit(docs)
```

```python
# Create the term-document matrix
# Transpose it so the terms are the rows
counts = count_vectorizer.transform(docs).transpose()
```

```python
counts.shape
```

```python
# Convert sparse matrix of counts to a gensim corpus
corpus = matutils.Sparse2Corpus(counts)
```

```python
id2word = dict((v, k) for k, v in count_vectorizer.vocabulary_.items())
```

```python
len(id2word)
```

```python
# Create lda model (equivalent to "fit" in sklearn)
lda = models.LdaModel(corpus=corpus, 
                      num_topics=20, 
                      alpha='asymmetric',
                      minimum_probability=0.03, 
                      id2word=id2word, passes=10)
```

```python
# lda.print_topics()
```

```python
# Transform the docs from the word space to the topic space (like "transform" in sklearn)
lda_corpus = lda[corpus]
lda_corpus
```

```python
# Store the documents' topic vectors in a list so we can take a peak
lda_docs = [doc for doc in lda_corpus]
```

```python
# Check out the document vectors in the topic space for the first 5 documents
for pres in zip(df['Name'], lda_docs):
# for pres in zip(df['Name'], df['Date'], lda_docs):
    print(pres)
```

```python
num_words = 30 # Number of words to include in word cloud
for t in range(lda.num_topics):
    plt.figure(figsize=(16,12))
    plt.imshow(WordCloud().fit_words(dict(lda.show_topic(t, num_words))))
    plt.axis("off")
    plt.title("Topic #" + str(t))
    plt.show()
```
