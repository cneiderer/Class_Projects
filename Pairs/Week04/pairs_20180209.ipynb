### Problem Description
There is a building with 100 floors. You have two eggs. If either egg is dropped from the **K**th floor or above it will break. If it's dropped from any of the floors below, it wouldn't break, however many times you dropped it. We need to find the cut-off floor **K**. How many tries would you need before you can find that floor?

```python
import numpy as np
```

```python
def find_floor(k):
    
    eggs = 2  
    floor = 3
    cnt = 0
    while eggs > 0: #floor != k:
        cnt += 1
        if floor < k:
            floor += 2
        elif floor > k:
            floor -= 1
            eggs -= 1
        else:
            if eggs == 2:
                floor += 2
            else:
                break                
            
    return cnt, floor      
```

```python
def find_floor2(k, m=1):
    
    eggs = m
    floor = eggs + 1
    cnt = 0
    while eggs > 0:
        cnt += 1
        if floor < k:
            floor += m
        elif floor > k:
            floor -= 1
            eggs -= 1
        else:
            if eggs == m:
                floor += m
            else:
                break                   
            
    return cnt, floor 
```
